### 데이터 필터링(불리언 색인)

```python
df['아파트'].unique()
```

```python
df[(df['아파트']=='호반현대')& (df['층']>5)]
```

```python
# 아파트 면적이 130이상인것만 필터링
#그중에 아파트 이름과 가격만 보고 싶다. 

df[df["면적"]>130][["아파트",'가격']]
```

```python
# 아파트 가격이 1.5억 미만인 아파트의 이름과 가격을 필터링
df[df['가격']<15000][["아파트",'가격']]
```

### 정렬

```python
# 아파트 중에서 가장 오래된 아파트 20개만 뽑아보자
df.sort_values(by='건축년도').head(20)# 오름차순
```

### 데이터 가공

```python
# 숫자데이터 가공하기 
# 면적과 가격을 이용해서 단가 컬럼을 생성하고 추가해보자 
df['단가']=df['가격']/df['면적']
```

```python
df.head()
```

```python
#지역별 아파트 평균 가격을 알아보자.
df['지역'].unique()
```

```python
def split_str(row):
    return row.split(" ")[0]
```

```python
df['가격'].mean()
```

```python
df['지역'].apply(split_str)
```

```python
df['지역2']=df['지역'].apply(split_str)
```

```python
df.head()
```

```python
#지역 2를 이용해서 광주 광역시의 평균 아파트 가격을 알아보자
```

```python
df[df['지역2']=='광주광역시']['가격'].mean()
```
