### numpy 불리언 인덱싱

- 특정 조건을 주어서 해당 조건에 맞는 값만 인덱싱 해오는 것을 말한다
- 조건이 맞으면 True, 아니면 False 값으로 배열을 만들어 조건에 맞는 값에 접근한다. 

```python
import numpy as np
```

```python
# 50~99까지의 랜덤한 8개의 값을 가지는 1차원 배열 생성 
arr20= np.array(np.random.randint(50,100, size=8))
arr20
```

```python
#boolean(불리언):True(참), 또는 False(거짓)으로 출ㄺ되는 자료형태
arr20>60
```

```python
#짝수 인 값들을 True로 출력시키기
arr20%2==0
```

```python
#arr20[조건식]
#실제 값을 보고 싶을 경우
#배열명[배열 안에서 찾고자하 하는 조건]
arr20[arr20%2==0]
```

```python
arr=np.array([[6,9,6,8,7],[2,3,5,4,1]])
```

```python
# 예제 
# 값이 홀수면 값을 덮어씌어 주기
num=0
arr[arr%2 !=0]=0
```

#### 4. bmi지수가 23이상 25이하인 사람은 몇명일까요?
- 불리언 인덱싱에서는 and-> &, or->|, nor->~을 쓴다.

```python
bmi[(bmi>=23) & (bmi<=25)]

```

```python
len(bmi[(bmi>=23) & (bmi<=25)])
```

```python
bmi[(bmi>=23) & (bmi<=25)].size
```

### 예제) 영화리뷰 평점 분석하기

#### 데이터 로드


```python
movie=np.loadtxt('data/ratings.dat',delimiter="::", dtype=np.int64)
movie
#사용자 번호, 영화 번호, 영화 평점, 입력시간
```

#### 2. 데이터 속성 확인
- 배열의 형태
- 배열 차원수
- 배열 전체 요소 개수

```python
print("배열의 형태:",movie.shape)
print("배열의 차원수:",movie.ndim)
print("배열의 요소 개수:",movie.size)
```

#### 3. 전체 영화의 평점 평균 구하기

```python
ratings=np.mean(movie[:,2])
```

```python
ratings
```

#### 4. 1번사용자가 매긴 평점의 평균 구하기
- 사용자 번호의 유일한 값(유니크 값) 확인
- 사용자 번호의 유니크 값 개수 확인
- 1번 사용자의 데이터에 접근
- 1번 사용자가 매긴 평점을의 평균 구하기


```python
# 평점을 매긴 사용자가 몇명인지 알아보자
user_id=np.unique(movie[:,0])
```

```python
user_id
```

```python
#사용자는 총 6040명
user_id.size
```

```python
#사용자의 번호가 1번인 데이터만 출력
one_user=movie[movie[:,0]==1]
one_user
```

```python
ratings=np.mean(one_user[:,2])
```

```python
ratings
```

#### 5. 각 사용자가 매긴 평덤들의 평균 구하기

```python
movie
```

```python
movie[movie[:,0]==1]
```

```python
movie[movie[:,0]==1][:,2].mean()
```

```python
user_mean_list=[]

for i in user_id: # range(1,6041,1)
    # 1~6040번 사용자가 매긴 평점들의 평균을 각각 구해주는 코드
    user_mean=movie[movie[:,0]==i][:,2].mean()
    #빈리스트에 사용자 번호와 그사람이 매긴 평점의 평균을 리스트로 저장
    user_mean_list.append([i,user_mean])

user_mean_list
```

#### 6. 위에서 구한 각 사용자별 평점 평균이 4 이상인 사용자만 구해보세요

- 사용자 번호와 명수를 구해봅시다


```python
# 불리언 인덱싱 적용을 위해 리스트를 배열로 변환
user_mean_array=np.array(user_mean_list, dtype='int64')
```

```python
user_mean_array
```

```python
ratings=user_mean_array[user_mean_array[:,1]>=4]
```

```python
ratings
```

```python
len(ratings)
```

```python
ratings[:,0]
```
