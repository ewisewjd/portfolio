### Matplotlib
- 데이터 및 분석 결과를 그래프나 차트로 시각화하기 위한 기본적인 라이브러리
- 데이터 분석 결과를 수치만으로 설명하는 것보다 시각적인 도구들을 활용하면 훨씬 더 직관적으로 이해가능!

#### 그래프 종류 참고 사이트
https://datastory1.blogspot.com/2017/12/matplotlib.html

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# pyplot은 matplotlib을 간단하고 쉽게 표현해주는 모듈
# 복잡하고 다양하게 표현하는 pylab도 있으나 pyplot에 비해 많은 정보를 입력해야 함
```

### plot차트 : 선 그래프

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x, y)   # 하나의 라인
plt.show()       # 그려진 라인들을 하나의 표로 표기
```

- Line style

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x, y, ls='-.')
plt.show()
```

- Line width

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x, y, ls='-.', lw=2)
plt.show()
```

- Marker

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x, y, ls='-.', lw=2, marker='o')
plt.show()
```

- Line color

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x, y, ls='-.', lw=2, marker='o', c='c')
plt.show()
```

다음 시각화 그래프와 같이 그려보세요~!

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x, y, ls='--', marker='o', c='green')
plt.show()
```

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x, y, ls=':', marker='D', c='red', mfc='b', ms=7)
plt.show()
```

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x, y, ls='--', lw=5, marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.show()
```

#### 차트 옵션

- 그림 범위 지정

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x, y, ls='--', lw=5, marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.xlim(-2,10)   # x축의 범위
plt.ylim(-3,12)   # y축의 범위

plt.show()
```

- 틱(tick) 설정
  - 차트나 그래프에서 축상의 위치 표시 지점을 틱(tick)이라고 함
  - 틱에 쓰여준 숫자 혹은 문자는 틱 라벨(tick label)이라고 함
  - 틱의 위치나 라벨은 matplotlib에서 자동으로 정해주지만 수동으로 설정하고 싶다면 xticks, yticks 명령을 사용!

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x, y, ls='--', lw=5, marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.xticks([0,3,6])
plt.yticks([1,5,9])

plt.show()
```

- 그리드(grid)
  - 그림의 틱 위치를 잘 보여주기 위해 차트 내부에 격자 선을 그려주는 것

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x, y, ls='--', lw=5, marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.grid()
plt.show()
```

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x, y, ls='--', lw=5, marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.grid()
plt.show()
```

#### 여러 개의 플롯을 한번에 그리기

```python
x = np.arange(4)
y = [2,4,6,8]
z = [3,6,9,12]

plt.plot(x, label='No.1')
plt.plot(y, label='No.2')
plt.plot(z, label='No.3')

# loc : 범례의 위치를 표시 (upper, center, lower), (left, center, right)
# prop={'size':5} : 범례의 크기 조절
plt.legend(loc='center', prop={'size':5})  # 범례 출력
plt.show()
```

```python
plt.plot(x)
plt.plot(y)
plt.show()

plt.plot(z)
plt.show()
```

#### bar chart : 막대 그래프

```python
x = np.arange(6)
y = [80,85,70,60,50,90]

plt.bar(x, y)
plt.show()
```

```python
# 막대 색상 변경
plt.bar(x, y, color='orange')
plt.show()
```

```python
plt.bar(x, y, color=['orange', 'cyan', 'pink'])
plt.show()
```

- 색상 종류 사이트 
https://matplotlib.org/3.2.1/gallery/color/named_colors.html

```python
# 막대 테두리 색상 설정
plt.bar(x, y, color=['orange', 'cyan', 'pink'], edgecolor='blue')
plt.show()
```

```python
# 막대 테두리 굵기 설정
plt.bar(x, y, color=['orange', 'cyan', 'pink'], edgecolor='blue', linewidth=4)
plt.show()
```

```python
# 투명도 설정
plt.bar(x, y, color=['orange', 'cyan', 'pink'], edgecolor='blue', linewidth=4, alpha=0.5)
plt.show()
```

```python
plt.bar(x, y, color=['orange', 'cyan', 'pink'], edgecolor='blue', linewidth=4, alpha=0.5)
plt.ylim(40,100)
plt.show()
```

```python
# x,y축 라벨 및 그래프 타이틀 설정
plt.bar(x, y, color=['orange', 'cyan', 'pink'], edgecolor='blue', lw=4, alpha=0.5)
plt.ylim(40,100)

plt.xlabel("No.")    # x축 라벨
plt.ylabel("Score")  # y축 라벨
plt.title("Exam")    # 그래프 타이틀

plt.show()
```
