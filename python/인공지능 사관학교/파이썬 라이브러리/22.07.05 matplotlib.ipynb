### plot차트: 선그래프


```python
x=[1,2,3,4,5]
y=[1,5,7,8,6]

plt.plot(x,y)#하나의 라인
plt.show() #그려진 라인들을 하나의 표로 표기
```

- line style

```python
x=[1,2,3,4,5]
y=[1,5,7,8,6]

plt.plot(x,y, ls='-.')#하나의 라인
plt.show() #그려진 라인들을 하나의 표로 표기
```

- line width

```python
x=[1,2,3,4,5]
y=[1,5,7,8,6]

plt.plot(x,y, ls='-.',lw=4)
plt.show()
```

- maker

```python
x=[1,2,3,4,5]
y=[1,5,7,8,6]

plt.plot(x,y, ls='-.',lw=2, marker='s')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,5,7,8,6]

plt.plot(x,y, ls='-.',lw=2, marker='o',c='c')
plt.show()
```

```python
x=np.arange(7)
y=[1,4,5,8,9,5,3]

plt.plot(x,y, ls='--',lw=4,marker='o',mew=5,mec='g',mfc='r',ms=15)
plt.show()
```

```python
x=np.arange(7)
y=[1,4,5,8,9,5,3]

plt.plot(x,y, ls='--',lw=4,marker='o',mew=5,mec='g',mfc='r',ms=15)
plt.xlim(-2,12) #x축의 범위
plt.ylim(-3,12) # y축의 범위
plt.show()
```

- 틱(tick)설정
    - 차트나 그래프에서 축상의 위치 표시 지점을 틱(tick)이라고 함
    - 틱에 쓰여준 숫자 혹은 문자는 틱 라벨(tick label)이라고 함
    - 틱에 위치나 라벨은 matplotlib에서 자동으로 정해주지만 수동으로 설정하고 싶다면 xticks, yticks를 사용

```python
x=np.arange(7)
y=[1,4,5,8,9,5,3]

plt.plot(x,y, ls='--',lw=4,marker='o',mew=5,mec='g',mfc='r',ms=15)
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.show()
```

- 그리드 (grid)
    - 그림의 틱 위치를 잘보여주기 위해 차트 내부에 격자 선을 그어주는 것

```python
x=np.arange(7)
y=[1,4,5,8,9,5,3]

plt.plot(x,y, ls='--',lw=4,marker='o',mew=5,mec='g',mfc='r',ms=15)
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.grid()
plt.show()
```

#### 사고 유형이 차대차 사건중애 죽거나 다친 사람이 많은 발생지 시도를 알아보고 시각화 해보자!

- 중요한 키워드 들이 들어위치 확인
    - 사고유형_ 대분류에서 '차대차' 행만 출력
    - '차대차' 인 행중에서 '발생지 시도'와 '사상자수' 컬럼확인

```python
data['사고유형_대분류'].value_counts()
```

```python
ac=data[data['사고유형_대분류']=='차대차']
```

```python
# 두 칼럼에 접근하려면 대괄호 안에 리스트로 접근
ac[['발생지시도','사상자수']]
```

```python
#사고 유형이 차대차인 사건의 총개수
len(ac)
```

```python
#발생한 1642개의 사건 중에서 시도 각각의 개수를 확인 
ac['발생지시도'].value_counts()
```

#### 각 발생지 시도별로 사상자수의 합계구하기

```python
ac2=ac[['발생지시도','사상자수']].groupby('발생지시도').sum()
ac2
```

```python
ac3=ac2.sort_values(by='사상자수',ascending=False)
ac3.index
```

```python
len(ac2.index)
```

```python
ac2.sort_values(by='사상자수', ascending=False)
```

```python
x=len(ac2.index)
x
```

```python
y=ac2.sort_values(by='사상자수', ascending=False)
```

```python
y=y['사상자수']
y
```

```python
# 그래프를 만들어 보자 
x=np.arange(17)


plt.bar(x,y)
#문자로 tick값들을 설정하기 위해서는 먼저 좌측부터 인덱스 번호로 개수를 설정해야함
plt.xticks(np.arange(17), labels=ac3.index)
plt.xlabel('지역')
plt.ylabel('사상자수')

plt.title('차vs차 교통사고의 사상자 수')
plt.show()
```

#### 예제 차vs차 교통사고 사상자수 시각화 하기
- x,y축 값 설정

```python
#x축에 들어갈 bar의 개수는 17개
x=np.arange(len(ac2.index))

#y축은 위에서 구한 발생지 시도별 사상자수 데이터를 내림차순 정렬로 넣어줌 
y=ac2['사상자수'].sort_values(ascending=False)
```

```python
y.index
```

- 그래프 그리기

```python
#차트 크기 변경 (가로,세로)
plt.figure(figsize=(13,7))
plt.bar(x,y)

plt.xticks(x, labels=y.index)
plt.xlabel('지역')
plt.ylabel('사상자 수')
plt.title('차vs차 교통사고의 사상자 수')
plt.show()
```
