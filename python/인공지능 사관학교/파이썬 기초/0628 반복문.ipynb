```python
#예제 4
# 랜덤하게 1~50사이의 숫자를 뽑으면 뽑은 숫자를 맞추는 게임
#숫자를 맞추면 접답을 맞추셧습니다 출력하고 입력을 멈추세요

import random
num=random.randint(1,50)

while True:
    my_num= int(input("숫자를 입력해주세요"))
    
    if my_num != num:
        if my_num<num:
            print(f"{my_num}보다 큰 숫자입니다")
        else:
            print(f"{my_num}보다 작은 숫자입니다")
            
    else:
        print("정답을 맞추셨습니다")
        break
```

```python
#답을 연속으로 맞출수록 점수가 커니는 ox 퀴즈가 있다 o은 문제를 맞춘것이고 x는 틀린것이다. 문제를 맞춘경우
#그문제의 점수는 그 문제까지 연속된 o의 개수가 된다. 

#변수설정
#1. 총점을 더하기 위한 변수
#2. 정답을 맞추었을때 o일경우 카운트 하기위한 변수
total=0
ok=0
#3. ox를 입력하기 위한 변수
ox=input("ox입력>>")

#반복문 만들기 

#1. 입력이 될때 o가 입력되면 1을 더해준다. 
#2. o가 연달아서 있으면 추가로 1을 더해준다
#3. x가 나온다면 0k를 초기화 시킨다.
for i in ox:
    if i=="o":
        ok+=1
        total+=ok
    else:
        ok=0
        
print(f"{total}점")
```

```python
# 숙제 문제 

#다음 score_list에 들어있는 데이터를 이용해 아래와 같이 성적표흘 만드는 프로그램으 작성하시오(단 학생의 수가 늘어나도 동작할수 있도록 고려하시오)

score_list=[["박매일",89,78,99],
           ["김미희",67,88,77],
           ["강예진",100,55,96]]

print("="*50)
print("이름\t","python\t","java\t","DB\t","평균\t","학점\t")
print("="*50)

python,java,db=0,0,0

for i in score_list:
    #선생님별 평균
    avg= (i[1]+i[2]+i[3])/3
    #과목별 평균
    avg_p, avg_j,avg_d=0,0,0
    
    if avg>=90:
        grade="A"
    elif avg>=80:
        grade='B'
    elif avg>=70:
        grade='C'
    elif avg>=60:
        grade='D'
        
    else:
        grade='F'
        
    #4~6번째 까지 출력하는 코드 
    print(i[0],"\t",i[1],"\t",i[2],"\t",i[3],"\t",int(avg),"\t",grade)

    
    #과목별 평균 알고리즘
    python+=i[1] #파이썬 점수의 합
    java+=i[2]  #java 점수의 합
    db+=i[3]   # db 점수의 합
    
    avg_p= python/len(score_list)
    avg_j=java/len(score_list)
    avg_d=db/len(score_list)
    
#문단 나누기
print("="*50)
# 맨 마지막 줄 출력 
print("평균\t",int(avg_p),"\t",int(avg_j),"\t",int(avg_d))
```

```python
# 네이버 영화 리뷰 분석
import pickle

#pickle:리스트와 같은 자료구조 자체를 저장하거나 불러오게 해주는 라이브러리(기능들의 집합)


#pickle모듈을 이용하여 데이터 로드

with open('data/movie_review.pkl','rb') as f:
    my=pickle.load(f)
```

```python
#선생님 코드
max_len=len(my[0][0])#첫번째 리뷰 길이를 넣어준다

max_content=""#가장 길이가 긴 리뷰가 들어갈 변수

for i in my:
    review_len= len(i[0])  #변수를 만들어서 변수에 리뷰 길이를 넣는다
    if max_len<review_len: #조건문을 만들어서 새롭게 들어오는 리뷰의 길이가 이전까지 리뷰길이보다 크다면
        max_len= review_len #max_len을 바꾼다
        max_content=i[0]    #리뷰값도 업데이트를 시켜준다
        
print(f"리뷰 최대길이는 {max_len}입니다.")
print(max_content)
```

```python
# 최소값 구하기

length=my[0]
for i,j in my:
    if len(i)<len(length):
        length=i
print(f"리뷰 최대길이는 '{len(length)}'입니다")
print(length)
```

```python
# 긍정리뷰와 부정리뷰의 개수를 파악하기
positive=0
nagative=0
for i,j in my:
    if j==1:
        positive+=1
    else:
        nagative+=1
        
print(f"긍정 '{positive}'개")
print(f"부정 '{nagative}'개")
```

```python
#선생님의 코드

neg_cnt=0 #부정개수
pos_cnt=0 #긍정개수

for i in my:
    if i[1]==0: #부정일경우
        neg_cnt+=1 
    else:        #긍정일경우
        pos_cnt+=1
        
print(f"긍정'{pos_cnt}'개")
print(f"부정'{neg_cnt}'개")
```
