```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/datagy/data/main/KNN_data.csv')
df.head()
df.plot()
```

```python
#label => large, medium, small 각각 df를 만드시오
small_df=df[df['Label']=='Small']
medium_df=df[df['Label']=='Medium']
large_df=df[df['Label']=='Large']

ax=small_df.plot(x='x',y='y', kind='scatter' ,color='r',label='small')
medium_df.plot(x='x',y='y', kind='scatter', color='g',label='medium',ax=ax)
large_df.plot(x='x',y='y', kind='scatter', color='b',label='large',ax=ax)
```

```python
small_df = df[df['Label'] == "Small"]
medium_df = df[df['Label'] == "Medium"]
large_df = df[df['Label'] == "Large"]

plt.scatter(small_df["x"], small_df["y"], color="r", label="Small")
plt.scatter(medium_df["x"], medium_df["y"], color="g", label="Medium")
plt.scatter(large_df["x"], large_df["y"], color="b", label="Large")

plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False

data = {
    '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
    '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
    '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
    '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
df
```

```python
## 1. 영화 데이터를 활용하여 x 축은 영화, y축은 평점인 막대 그래프를 만드시오.
plt.figure(figsize=(20,8))
plt.bar(df['영화'],df['평점'])
```

```python
## 2. 앞에서 만든 막대 그래프에 제시된 세부 사항을 적용하시오.
#- 제목: 국내 Top 8 영화 평점 정보
#- x축 label: 영화 (90도 회전)
#- y축 label: 평점
plt.figure(figsize=(20,8))
plt.bar(df['영화'],df['평점'])
plt.title('국내 top 8 영화 평점 정보')
plt.xlabel("영화")
plt.xticks(rotation=90)
plt.ylabel("평점")
```

```python
## 2. 앞에서 만든 막대 그래프에 제시된 세부 사항을 적용하시오.
#- 제목: 국내 Top 8 영화 평점 정보
#- x축 label: 영화 (90도 회전)
#- y축 label: 평점
plt.figure(figsize=(20,8))
plt.bar(df['영화'],df['평점'])
plt.title('국내 top 8 영화 평점 정보')
plt.xlabel("영화")
plt.xticks(rotation=90)
plt.ylabel("평점")
```

```python
## 3. 개봉 연도별 평점 변화 추이를 꺾은선 그래프로 그리시오.
#연도별 평균 데이터를 구하는 코드는 다음과 같습니다.
#df_group = df.groupby('개봉 연도').mean()

df_group = df.groupby('개봉 연도').mean(numeric_only=True
)
df_group

plt.plot(df_group.index, df_group['평점'])
```

```python
## 3. 개봉 연도별 평점 변화 추이를 꺾은선 그래프로 그리시오.
#연도별 평균 데이터를 구하는 코드는 다음과 같습니다.
#df_group = df.groupby('개봉 연도').mean()

df_group = df.groupby('개봉 연도').mean(numeric_only=True
)
plt.plot(df_group.index, df_group['평점'])
```

```python
## 4) 앞에서 만든 그래프에 제시된 세부 사항을 적용하시오.
#- marker: 'o'
#- x축 눈금: 5년 단위 (2005, 2010, 2015, 2020)
#- y축 범위: 최소 7, 최대 10
plt.plot(df_group.index, df_group['평점'], marker='o')
plt.xticks([2005,2010,2015,2020])
plt.ylim(7,10)
```

```python
## 4) 앞에서 만든 그래프에 제시된 세부 사항을 적용하시오.
#- marker: 'o'
#- x축 눈금: 5년 단위 (2005, 2010, 2015, 2020) (xticks)
#- y축 범위: 최소 7, 최대 10

plt.plot(df_group.index, df_group['평점'], marker='o')
plt.xticks([2005, 2010, 2015, 2020])
plt.ylim(7,10)
```

```python
## 5) 평점이 9점 이상인 영화의 비율을 확인할 수 있는 원 그래프를 제시된 세부 사항을 적용하여 그리시오.
#- label: 9점 이상 / 9점 미만
#- 퍼센트: 소수점 첫 번째 자리까지 표시
#- 범례: 그래프 우측에 표시


df[df['평점']>=9.0]
df[df['평점']<9.0]

values=[len(df[df['평점']>=9.0]),len(df[df['평점']<9.0])]
labels=["평점 9점이상","평점 9점 이하"]

plt.pie(values, labels=labels, autopct="%1.f%%")
plt.legend(loc=(1,0.3))

plt.show()
```

```python
#matplotlib를 사용하여 다중 서브플롯(multi subplot)을 그려보세요. 
#1행 2열의 구조에서, 왼쪽 서브플롯에는 0부터 10까지의 범위에서 sin(x) 함수를, 
#오른쪽 서브플롯에는 0부터 10까지의 범위에서 cos(x) 함수를 그립니다.

x = np.linspace(0, 10, 100)

y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8,4))

ax[0].plot(x, y1)
ax[1].plot(x, y2)

plt.show()
```
