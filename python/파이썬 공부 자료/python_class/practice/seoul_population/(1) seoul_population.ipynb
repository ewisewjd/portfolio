# 서울시 인구 데이터 분석

```python
# 라이브러리 import
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 데이터 파일 read
df = pd.read_excel('seoul_popular_data.xls', header=1, usecols="B, D, G, J, N")
df
```

```python
# 컬럼명 변경:
# (자치구 -> 구, 합계 -> 인구수, 등록외국인 -> 외국인, 65세이상고령자 -> 고령자)

df.columns = ['구', '인구수', '한국인', '외국인', '고령자']
df
```

```python
# 합계행 제거
df.drop([0], inplace=True)
df
```

```python
df.drop([1], inplace=True)
df
```

```python
new_df = df.reset_index()
new_df
```

```python
new_df.drop('index', inplace=True, axis=1)
new_df
```

```python
## (4) 각 구별 전체 인구를 이용해서 구별 '외국인비율'과 '고령자비율' 컬럼을 추가하고, 소수점 첫째자리 반올림하기
### - 외국인 비율: 외국인 / 인구수 * 100
### - 고령자 비율: 고령자 / 인구수 * 100

new_df['인구수'] = new_df['인구수'].astype('int')
new_df['한국인'] = new_df['한국인'].astype('int')
new_df['외국인'] = new_df['외국인'].astype('int')
new_df['고령자'] = new_df['고령자'].astype('int')

new_df.info()
```

```python
# 외국인비율 = 외국인 / 인구수 * 100
new_df['외국인비율'] = round((new_df['외국인'] / new_df['인구수']) * 100, 1)
new_df['외국인비율']
```

```python
# 고령자비율 = 외국인 / 인구수 * 100
new_df['고령자비율'] = round((new_df['고령자'] / new_df['인구수']) * 100, 1)
new_df['고령자비율']
```

```python
# (5) 인구수별로 내림차순 정렬(sort)하기
new_df.sort_values(by="인구수", ascending=False)
```

```python
# (6) 외국인, 외국인비율로 내림차순 정렬하기
new_df = new_df.sort_values(by='외국인', ascending=False)
new_df = new_df.sort_values(by='외국인비율', ascending=False)
new_df
```

```python
# (7) 고령자수 / 고령자비율로 내림차순 정렬하기
new_df1 = new_df.sort_values("고령자", ascending=False)
new_df2 = new_df.sort_values("고령자비율", ascending=False)

display(new_df1, new_df2)
```

```python
# (8) 인덱스 값을 '구'별 데이터로 변경하기 - groupby 활용
visual_df = new_df.set_index('구')
```

```python
# (9) 구별 인구수 데이터를 바 그래프로 내림차순 정렬 후 시각화하기
plt.rcParams["font.family"] = "Malgun Gothic" # Windows
plt.rcParams['font.family'] = "AppleGothic" # Mac
visual_df['인구수'].sort_values().plot(kind='barh', grid=True)

df.sort_values(by='인구수')
df['인구수'].sort_values()
```

```python
x = visual_df.index
y = visual_df['인구수']

plt.figure(figsize=(16,10))
plt.title("서울시 구별 인구 현황")
plt.xlabel("gu")
plt.ylabel("population")
plt.grid()
plt.bar(x,y)
plt.savefig("서울시인구현황.png")
```
