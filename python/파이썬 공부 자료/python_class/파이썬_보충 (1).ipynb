```python
def div(a: float, b: float, eps=0.0, debug=False) -> float:
    '''
    input :  a, b
    return : a / b
    '''
    if debug:
        print("a, b : ", a, b)
    if eps: # True
        N = a / (b + eps)
        if debug:
            print("N : ", N)
        return N
    else:  # False
        try:
            N = a / b  # epslion를 안쓰는 경우.
            return N
        except ZeroDivisionError:
            print("b is 0.")

div(3, 0, debug=True)
div(3, 0, 1e-6, True)
div(debug=True, eps=1e-3, b=1, a=10)
```
