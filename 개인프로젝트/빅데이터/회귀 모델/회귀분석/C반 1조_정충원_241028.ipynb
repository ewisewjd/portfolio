## 회귀 모델 생성

```python
## 먼저 데이터를 분리한다. train과 test로 나누었다. 
df_train,df_test=train_test_split(df_DIABETES,test_size=0.3,random_state=1234)
print(f"train_data_size{df_train.shape}")
print(f"test_data_size{df_test.shape}")
```

```python
## 선형 회귀 분석 
reg_model=smf.ols(formula='Y~AGE+GENDER+BMI+BP+CHOLESTEROL+LDL+HDL+TCH+LTG+GLUCOSE',data=df_train)

#적합
reg_result=reg_model.fit()
print(reg_result.summary())
```

- 선형 회귀분석을 실시한다. 이때 모든 변수들에 대해 실시하여본다. 

- 위 결과를 실시하여 본바 다음과 같다,
    - F분석 결과 값은 0.05를 기준으로 아주 작으므로 회귀모델로서 유의미한 계수가 존재한다. 
    - 결정계수 R^2의 값 특히 수정 결정계수의 값은 0.511이다. 이는 이 회귀모델이 약 50정도의 설명력을 갖는다는 것을 의미한다고 할수 있다. 
    - 각 변수들을 보자면 일부 값들이 0.00수치를 기록하여 유의수준보다 작으므로 변수로써 유의하다고 볼수있다. 

## 다중 공산성 확인 

```python
# 먼저 데이터에서 목표변수를 제거한다. 
df_train_x=df_train.drop('Y',axis=1)


## statsmodel의 상수항 추가함수 적용 
df_train_x_const=add_constant(df_train_x)

## 데이터 프레임으로 저장
df_vif=pd.DataFrame()
df_vif['variable']=df_train_x_const.columns

df_vif['VIF']=[variance_inflation_factor(df_train_x_const.values,i)for i in range(df_train_x_const.shape[1])]


#df_vif정렬
df_vif.sort_values('VIF',ascending=True).round(3)
```

- 다중 공산성을 확인해 본결과 10을 초과하는 변수 5개를 발견하였다.
- 그중 콜레스트롤 변수는 ldl(저밀도 콜레스트롤)과 상관관계가 높고 선형성을 갖추고 있으므로 제거를 하는 것이 좋을것같다고 판단된다. 
    - 먼저 저밀도 콜레스테롤의 경우 심혈관을 유발하는 일명 나쁜 콜레스트롤로 분류된다. 따라서 당뇨병을 예측하는데 더 효과적일 수 있다. 
    - 두번째로 콜레스테롤의 다중 공산성 수치가 훨씬 높다. 이는 콜레스테롤과 lpl이 중복되는 부분이 많다는 뜻이며 자칫 변수들의 영향력을 해칠 가능성이 있다. 

## 변수 제거후 재 학습

```python
## 선형 회귀 분석 
reg_model=smf.ols(formula='Y~AGE+GENDER+BMI+BP+LDL+HDL+TCH+LTG+GLUCOSE',data=df_train)

#적합
reg_result=reg_model.fit()
print(reg_result.summary())
```

- 콜레스트롤을 제거한 후 다시 모델을 학습하였다. 
- f검정의 결과는 훨씬 낮아졌고 이는 모델의 유의미함이 증가하였음을 나타낸다. 
- 반면 수정 결정계수는 0.01차이로 낮아졌다. 이는 모델의 설명력에 큰 변화가 없음을 나타낸다.
- 각 값들의 p밸류 값들이 조정되었다. 
    - 일반적으로 생각하자면 나이는 당뇨병의 유무와 큰 영향을 미칠것으로 보였다 그러나 p밸류값을 기준으로 분석결과 나이는 당뇨병의 유무와 유의미한 영향을 주지 못했다. 
    - 기존 저밀도 콜레스트롤의 p밸류값은 콜레스트롤의 제거로 인해 낮아졌다. 그러나 여전히 0.05이상의 값을 초과하고 있다. 
    - p밸류 값을 기준으로 한다면 bmi와 bp,ltg,hdl등이 변수로서 유의미함을 갖춘다고 볼수 있다.

- 정규분포 직선을 중심으로 따르므로 정규분포를 따른다고 볼 수 있다.

- 잔차의 독립성을 확인해본 결과 잔차가 관리 상하한을 벗어나지 않고 랜덤하게 분포되어 있으므로 독립성을 만족

## 후진제거법을 통한 변수 셀렉션


```python
## 후진 제거법을 이용한 변수 선택

## 목표변수와 설명변수 분리 
df_train_y=df_train['Y']
df_train_x=df_train.drop('Y',axis=1)

#sklearn의 선형회귀 모델
model=LinearRegression()

##전체 변수 중 5개의 변수가 남을 때까지 변수 선택
rfe=RFE(estimator=model, n_features_to_select=5).fit(df_train_x,df_train_y)

## 선택된 변수 
selected_cols=df_train_x.columns[rfe.support_]

#제거된 변수 
removed_cols=df_train_x.columns[~rfe.support_]


## 선택된 변수 
print(f"Selected Variable {selected_cols}")
print(f"Removed Variable {removed_cols}")
```

```python
# 선택된 변수 개수 확인
print(f"Selected Variables (count: {len(selected_cols)}): {selected_cols}")
print(f"Removed Variables (count: {len(removed_cols)}): {removed_cols}")
```

```python
# 후진 제거법을 이용한 변수 선택 및 모델 생성 
# 회귀 모델생성 
rfe_reg_model=smf.ols(formula='Y~GENDER+BMI+BP+TCH+LTG', data=df_train)

#적합
rfe_reg_result=rfe_reg_model.fit()
print(rfe_reg_result.summary())
```

- 후진 선택법은 입력된 모델등을 토대로 변수 중요도에 따라 변수들을 하나씩 제거하는 방법이다. 
- 위 방법을 토대로 회귀식을 돌려본 결과 
    - 결정계수는 0.49로 조금 낮아졌다. 
    - TCH의 값을 제외한 다른 변수들의 P값들이 유의한 값임을 나타낸다. 
    - TCH는 헤모글로빈을 나타냄으로 이 변수의 값은 당뇨에 유의미한 영향을 주지 못하는것으로 보인다. 
- 또한 후진 선택법으로 선택한 결정계수가 다중공선성과 일반선택에 의한 변수의 결정계수보다 낮으므로 이 방식이 언제나 유용한것은 아닌것 처럼 보인다. 


```python
# 절편을 제외한 회귀 계수만 사용
df_reg_coef = pd.DataFrame({"Coef": rfe_reg_result.params.values[1:]}, index=['GENDER', 'BMI', 'BP', 'TCH', 'LTG'])

# 수평 막대 그래프 그리기
df_reg_coef.plot.barh(y="Coef")
```

- 위 그래프는 회귀꼐수를 나타낸 값이다. GENDER와 LTG가 가장 큰 회귀 계수를 가진다.
- 즉 성별과 LTG의 값이 당뇨병에 큰 영향을 미친다고 볼수있다.

# 실습 2. 규제화 

```python
## 패키지 불러오기 
from sklearn.linear_model import Ridge,Lasso,ElasticNet
```

```python
## 목표변수 설명변수 데이터 지정 하기 

df_DIABETES_y=df_DIABETES['Y']
df_DIABETES_x=df_DIABETES.drop('Y',axis=1, inplace=False)

# train/ test data 분리 
df_train_x,df_test_x,df_train_y,df_test_y=train_test_split(df_DIABETES_x,df_DIABETES_y, test_size=0.3, random_state=1234)

print(f"train data x size: {df_train_x.shape}")
print(f"train data x size: {df_train_y.shape}")
print(f"train data x size: {df_test_x.shape}")
print(f"train data x size: {df_test_y.shape}")
```

```python
### 모델 생성 
ridge=Ridge(random_state=1234, alpha=1)
ridge.fit(df_train_x,df_train_y)

#절편값 확인 
print(ridge.intercept_)

# 회귀계수 확인 

df_ridge_coef=pd.DataFrame({"Coef":ridge.coef_},index=df_train_x.columns)
df_ridge_coef

## 회귀계수 그래프 
df_ridge_coef.plot.barh(y="Coef",legend=False)
```

```python
### 규제화회귀 모델 생성 : ridge 모델
###모델의 성능 
train_pred=ridge.predict(df_train_x)
test_pred=ridge.predict(df_test_x)

print(f"train data 결정계수:{r2_score(df_train_y,train_pred)}")

print(f"test data 결정계수:{r2_score(df_test_y,test_pred)}")
```

```python
### 모델 생성 lasso
lasso=Lasso(random_state=1234, alpha=1)
lasso.fit(df_train_x,df_train_y)

#절편값 확인 
print(lasso.intercept_)

# 회귀계수 확인 

df_lasso_coef=pd.DataFrame({"Coef":lasso.coef_},index=df_train_x.columns)
df_lasso_coef

## 회귀계수 그래프 
df_lasso_coef.plot.barh(y="Coef",legend=False)
```

- 규제화는 회귀계수가 가질 수 있는 값에 대한 제약조건을 부여하여 모델의 분산을 감소시키고 모델의 일반화(안정성)성능을 높인다. 

- 그러나 RIDGE, LASSO, ELASTIC 셋의 성능을 비교하여 본 결과 큰 차이가 있지는 않았다. 
- 마찬가지로 세가지 규제화 방식의 경우에도 GENDER와 LTG가 가장 회귀계수가 높았다.
