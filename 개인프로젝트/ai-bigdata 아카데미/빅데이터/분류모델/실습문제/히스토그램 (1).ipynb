```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

# 그래프 한글폰트 적용:나눔 고딕
matplotlib.rc("font", family = "AppleGothic")
# 그래프 (-) 기호 표시
matplotlib.rc("axes", unicode_minus = False)

df_raw = pd.read_csv('/Users/CKY/Desktop/01 실습파일/2. Big Data 분석/SCALE불량.csv', encoding='euc-kr')
df_raw
```

```python
df_raw['scale'] = np.where(df_raw['scale']=='양품',0,1)
df_raw
```

```python
df_raw[df_raw['scale'] == 1]
```

```python
# scale 값이 0인 데이터와 1인 데이터 분리
df0 = df_raw[df_raw['scale'] == 0]
df1 = df_raw[df_raw['scale'] == 1]

# 히스토그램 그리기
plt.figure(figsize=(5, 3))
plt.hist(df0['pt_thick'],alpha=0.5,color='green')
plt.hist(df1['pt_thick'],alpha=0.5,color='red')
plt.title('pt_thick')
```

두께가 얇을수록 scale 발생 확률이 올라간다.
