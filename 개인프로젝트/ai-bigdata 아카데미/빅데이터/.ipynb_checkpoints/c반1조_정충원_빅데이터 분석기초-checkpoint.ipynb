# 실습 7.

```python
df=pd.read_csv("./FITNESS.csv",encoding="euc-kr")
df.head()
```

```python
df_groupby=df.groupby(['GENDER','AGEGROUP'],as_index=False).mean()
```

```python
df_groupby
```

```python
df_groupby1=df_groupby.pivot(index='GENDER',columns='AGEGROUP',values='WEIGHT')
df_groupby1
```

```python
df_groupby2=df_groupby.pivot(index='AGEGROUP',columns='GENDER',values='WEIGHT')
df_groupby2
```

- 기억에 남았던 부분
    - 수업중 : "우리는 코딩을 하는것이 아니라 데이터를 이해하는것이 더 중요하다"-> 코딩은 단순히 데이터를 조작하기 위해 공부하는 것이다.
    - 그럼에도 불구하고 데이터를 적절하게 가공하고 우리가 원하는 결과를 얻어내려면 이를 능숙하게 다루는 것이 필요하다. 
    
- 실습 문제를 풀면서 다음과 같은 카테고리를 배울수 있었다. 
    - 데이터 생성 
    - 데이터 읽고 쓰기 
    - 데이터 정보 확인하기 
    - 데이터 선택하기 
    - 데이터 변환하기
    
- 궁금한 부분을 확인하였다.
    - loc는 인덱스의 라벨값으로 결과를 추출한다. 그렇다면 연속된 숫자가 아닌 라벨값도 :(콜론)을 이용해서 추출될까?
    - agg(['mean'])과 mean()는 왜 다른 값을 반환하는 것일까 
    - join 시에 on 옵션의 값이 키값이라면 그것을 index로 하여 합쳐지는가?
