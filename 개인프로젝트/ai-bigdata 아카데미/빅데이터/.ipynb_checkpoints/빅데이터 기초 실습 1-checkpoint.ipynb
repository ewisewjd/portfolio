### 데이터 선택


```python
# 공통 모듈 불러오기
import pandas as pd
import numpy as np
from numpy.random import randn
```

```python
np.random.seed(1234)
```

```python
df=pd.DataFrame(randn(10,2),columns=['A','B'])
df
```

#### slice와 loc,iloc를 이용한 선택

```python
# slice를 이용한 데이터 선택
df['A']
##시리즈 형태로 나온다
```

```python
df[['A']] #데이터 프레임 형태로 나온다.
```

```python
df.loc[3:8] #행선택
```

```python
df.loc[3:3] #단일 행선택
```

```python
#iloc를 이용한 선택
##iloc 열과 행을 index번호로 지정가능

df.iloc[0:7,[0]]
```

```python
#열선택 데이터 프레임으로 반환
df.iloc[:,[0]]
```

```python
#행선택
df.iloc[3:8]
```

```python
#단일 행 선택 
df.iloc[3:4]
```

```python
#단일 행 선택 - 시리즈로 반환
df.iloc[3]
```

```python
#query를 이용한 데이터 선택 
df.query('A>0.5')
```

```python
#query와 동일한 slice표현
df.query('A>B')
df[df.A>df.B]
```

```python
### query를 이용한 복수 선택 복수조건
### AND조건 
df.query('A>0.5 & B<0.0')
```

```python
#OR 조건
df.query('A> 0.5 | B<0.0')
```

```python
##질문
df.index=["가","나","다","라","마","바","사",",아","자","차"]
```

```python
df["가":"다"]
```

#### filter와 mask를 이용한 데이터 선택 

```python
df=pd.DataFrame({"ABC":['A','B','C','A'],
                "abc":['a','b','c','c'],
                'group1':['G1','G2','G3','G4'],
                'value1':[11,13,12,14],
                'val1_mod':[0.231,0.197,0.213,0.199]})

df
```

```python
df.filter(items=['group1','value1'])
```

```python
df.filter(regex='1',axis=1)
```

```python
df.filter(regex='1$',axis=1)
```

```python
#filter사용
df.filter(like='a',axis=1)
```

```python
#regex는 정규 표현식이다. 나중에 정규 표현식 공부하자 
df.filter(regex='[Aa]',axis=1)
```

#### 데이터 조작 실습

```python
import pandas as pd
import numpy as np

pop_seoul=pd.read_csv("./서울인구통계.txt",sep="\t",thousands=",",skiprows=2, encoding="utf-8")
pop_seoul.head()
```

```python
## 데이터 불러오고 구조 파악하기
pop1=pop_seoul[['자치구','계','계.1','계.2','65세이상고령자']]
pop_seoul=pop1.drop([0])
```

```python
pop_seoul.head()
```

```python
## 컬럼명을 구분할수 있도록 리네임
pop_seoul.rename(columns={pop_seoul.columns[0]:"구별",
                         pop_seoul.columns[1]:"인구수",
                         pop_seoul.columns[2]:"한국인",
                         pop_seoul.columns[3]:"외국인",
                         pop_seoul.columns[4]:"고령자"}, inplace=True)
pop_seoul.head()
```

```python
## 데이터 불러오고 구조 파악하기2
cctv_seoul=pd.read_excel("./서울시CCTV.xlsx", skiprows=2, names=['기관명','소계','2011년이전','2012년','2013년','2014년','2015년','2016년','2017년','2018년','2019년','2020년','2021년'])
```

```python
cctv_seoul.head()
```

```python
#컬럼 확인 
cctv_seoul.columns
```

```python
## 기관 명을 구별로 인덱스 이름 변경 
cctv_seoul.rename(columns={cctv_seoul.columns[0]:'구별'},inplace=True)
cctv_seoul.rename(columns={cctv_seoul.columns[1]:'cctv'},inplace=True)

cctv_seoul.head()
```
