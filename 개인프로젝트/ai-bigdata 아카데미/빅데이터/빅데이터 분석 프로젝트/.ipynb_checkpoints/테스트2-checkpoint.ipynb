```python
# 각 유저별로 가장 많이 시킨 상품을 찾기
def get_most_frequent_item(user_group):
    # 각 유저별로 상품 시킨 횟수 계산
    value_counts = user_group['ItemName'].value_counts()
    # 가장 많이 시킨 상품과 그 횟수만 반환
    return value_counts.idxmax(), value_counts.max()

# 그룹화하여 유저별로 가장 많이 시킨 상품 추출
most_frequent_items = df_raw.groupby('idUser').apply(get_most_frequent_item)

# 결과 확인
print(most_frequent_items)
```

```python
import pandas as pd
import scipy.stats as stats

# (1) 멤버십 가입자와 비가입자의 매출액 데이터 준비
df_member_sales = df_raw[df_raw['MemberYN'] == 'Y']['Price']
df_non_member_sales = df_raw[df_raw['MemberYN'] == 'N']['Price']

# (2) 두 집단 간의 평균 차이가 통계적으로 유의미한지 검정
t_stat, p_value = stats.ttest_ind(df_member_sales, df_non_member_sales)

# (3) 결과 출력
print("t-statistic:", t_stat)
print("p-value:", p_value)

# (4) p-value에 따른 결론 도출
alpha = 0.05  # 5% 유의수준

if p_value < alpha:
    print("귀무 가설을 기각합니다. 멤버십 가입자와 비가입자 간의 매출액 차이는 통계적으로 유의미합니다.")
else:
    print("귀무 가설을 채택합니다. 멤버십 가입자와 비가입자 간의 매출액 차이는 통계적으로 유의미하지 않습니다.")
```

```python
item_code_mapping = dict(zip(df_orders['ItemCode'], df_items['ItemName']))

# 유저별로 구매한 상품 코드 리스트 생성
item_code_list_per_user = df_orders.groupby(['idUser'])['ItemCode'].apply(list)

# 각 유저의 상품 코드 리스트를 상품 이름 리스트로 변환
item_name_list_per_user = item_code_list_per_user.apply(lambda x: [item_code_mapping.get(code, 'Unknown') for code in x])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter


# 한 명의 유저 (예: 'U10001')에 대해 빈도수 그래프 그리기
user_id = 'U10001'  # 원하는 유저의 id
item_names = item_name_list_per_user[user_id]

# 빈도수 계산
item_counts = Counter(item_names)

# 데이터프레임으로 변환
item_counts_df = pd.DataFrame(item_counts.items(), columns=['ItemName', 'Count'])
item_counts_df = item_counts_df.sort_values(by='Count', ascending=False)

# 그래프 그리기
plt.figure(figsize=(12, 6))
sns.barplot(x='Count', y='ItemName', data=item_counts_df, palette="viridis")
plt.title(f"Top Purchased Items by User {user_id}")
plt.xlabel("Frequency")
plt.ylabel("Item Name")
plt.tight_layout()
plt.show()
```

```python


# 1. 아이템 코드에서 중분류 추출하는 함수 정의
def extract_category(item_code):
    # 아이템 코드에서 'M17'과 같은 중분류 추출
    return item_code.split('-')[1]

# 2. 유저별로 아이템 코드에서 중분류 추출하고 빈도수 계산
user_id = 'U10001'  # 예시 유저
item_codes = item_code_list_per_user[user_id]

# 중분류 추출
categories = [extract_category(code) for code in item_codes]

# 빈도수 계산
category_counts = Counter(categories)

# 데이터프레임으로 변환
category_counts_df = pd.DataFrame(category_counts.items(), columns=['Category', 'Count'])
category_counts_df = category_counts_df.sort_values(by='Count', ascending=False)

# 3. 그래프 그리기
plt.figure(figsize=(12, 6))
sns.barplot(x='Count', y='Category', data=category_counts_df, palette="viridis")
plt.title(f"Top Purchased Categories by User {user_id}")
plt.xlabel("Frequency")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```

```python
# 여러 유저에 대해 카테고리 빈도수 계산
all_categories = []
for user_id, item_codes in item_code_list_per_user.items():
    categories = [extract_category(code) for code in item_codes]
    all_categories.extend(categories)

# 빈도수 계산
category_counts = Counter(all_categories)

# 데이터프레임으로 변환
category_counts_df = pd.DataFrame(category_counts.items(), columns=['Category', 'Count'])
category_counts_df = category_counts_df.sort_values(by='Count', ascending=False)

# 4. 그래프 그리기 (여러 유저에 대한 카테고리 빈도수 시각화)
plt.figure(figsize=(12, 6))
sns.barplot(x='Count', y='Category', data=category_counts_df, palette="viridis")
plt.title("Top Purchased Categories by All Users")
plt.xlabel("Frequency")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```

```python

# 1. 아이템 코드에서 중분류 추출하는 함수
def extract_category(item_code):
    return item_code.split('-')[1]  # 예: 'L4-M17-S0530-1024' -> 'M17'


# 3. 유저별, 연도별, 카테고리 빈도수 계산
category_counts_by_year = []

# 각 유저에 대해 연도별로 카테고리 추출 후 빈도수 계산
for user_id, group in df_raw.groupby('idUser'):
    for year, group_year in group.groupby('d_year'):
        categories = [extract_category(code) for code in group_year['ItemCode']]
        category_counts = Counter(categories)
        for category, count in category_counts.items():
            category_counts_by_year.append({
                'UserID': user_id,
                'Year': year,
                'Category': category,
                'Count': count
            })

# 4. 카테고리 빈도수 데이터프레임으로 변환
category_counts_by_year_df = pd.DataFrame(category_counts_by_year)

# 5. 연도별로 카테고리 빈도수 계산
category_counts_by_year_summary = category_counts_by_year_df.groupby(['Year', 'Category'])['Count'].sum().reset_index()

# 6. 그래프 그리기
plt.figure(figsize=(12, 6))
sns.barplot(x='Count', y='Category', hue='Year', data=category_counts_by_year_summary, palette='viridis')

plt.title("Top Purchased Categories by Year")
plt.xlabel("Frequency")
plt.ylabel("Category")
plt.legend(title="Year")
plt.tight_layout()
plt.show()
```
